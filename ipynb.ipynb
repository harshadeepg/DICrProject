```r
generateSquares <- function(x){
    return(x^2)
}
```

```r
a = generateSquares(4)
print(a)
```

```r
generateSquares2 <- function(x){
    for(i in 1:x){
        y <- i^2
        print(y)
    }
    return(x^2)
}
```

```r
print(generateSquares2(4))
```

```r
sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2 <- rpois(12,34)
```

```r
par(bg="cornsilk")
```

```r
for (i in 1:5){
    plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="month", ylab= "Sales")
    title(main="sales by Month")
    lines(sales2, type="o", pch=22, lty=2, col="red")
    grid(nx=NA, ny=NULL)
    legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
    name = paste(c('filename',i), collapse = '_')
    filename = paste(c(name, '.pdf'))
    dev.print(pdf, filename)
}
```
